```python
import pandas as pd
```

```python
homicides = pd.read_csv('Homicide.csv')
```

```python
#how many rows are in the dataset
homicides.shape[0]
```

```python
#list the columns
homicides.columns
```

```python
#replace the default index with the values of the 'Index_' column
homicides.set_index('Index_', inplace=True)
homicides
```

```python
#list the types of homicides in the order of their frequency (most to least)
homicides['Homicide_Type'].value_counts()
```

```python
#find the year with the highest amount of shootings
homicides[homicides['Homicide_Type'] == 'Shooting']
```

```python
#which neighbourhood had the highest amount of shootings
homicides[homicides['Homicide_Type'] == 'Shooting']['Neighbourhood'].value_counts().head(10)
```

```python
#list the 5 safest neighbourhoods
homicides['Neighbourhood'].value_counts().tail()
```

```python
homicides.groupby('Homicide_Type').agg('count')
```

```python
#show on a map, your five safest neighbourhoods
import matplotlib.pyplot as plt
homicides.plot(kind="scatter", x="X", y="Y", alpha=1)
plt.show()
```

```python
#plot the dataset on a map
import matplotlib.pyplot as plt
homicides.plot(kind="scatter", x="Long", y="Lat", alpha=1)
plt.show()
```
