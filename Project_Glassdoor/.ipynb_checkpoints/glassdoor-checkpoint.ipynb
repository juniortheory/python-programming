Used Selenium Python library to scrape data from the Glassdoor website. Selenium automates a webbrowser. We use chromedriver to automate Google Chrome webbrowser 

```python
import time
import csv
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.select import Select
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.chrome.options import Options


```

```python
#initalizes the chrome driver. added options for window browse and broswer visibility  
class Browser:    
    def __init__(self):
        self.chrome_options = Options()
        #self.chrome_options.add_argument("--headless") # hide browser
        self.chrome_options.add_argument("--window-size=1920x1080")
        self.chrome_options.add_argument("user-agent=Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.102 Safari/537.36")
        self.chrome_options.add_argument("--start-maximized")
        self.path_to_chromedriver = './chromedriver 2' # change path as needed
        self.browser = webdriver.Chrome(executable_path = self.path_to_chromedriver, chrome_options = self.chrome_options)
    
#use_browser is used to access the web broswer    
    def use_browser(self):
        return self.browser
```

```python
browser = Browser().use_browser()
```

```python
#goes to Glassdoor website
browser.get("https://glassdoor.ca")
```

```python
#finding the element in the job search field
keyword_field = browser.find_element_by_xpath("//input[@id='KeywordSearch']")
location_field = browser.find_element_by_xpath("//input[@id='LocationSearch']")
search_button = browser.find_element_by_xpath("//button[@id='HeroSearchButton']")
```

```python
#user input for job search and location
keyword = input("Job Title, Keywords or Company: ")
location = input("Location: ")
```

```python
#clears search field
keyword_field.clear()
keyword_field.send_keys(keyword)
location_field.clear()
location_field.send_keys(location)
search_button.click()
```

```python
#reads page  and each pagenumbers of job search
url_format =  browser.current_url
url_format = url_format.replace(".htm", "_IP{}.htm")
```

```python
#scraping data from every every job posting
def get_job_info(url, job_id):
    browser.get(url)
    if "ad.doubleclick.net" in browser.current_url:
        browser.get("https://" + browser.current_url.split("https://",2)[2])
        
    print("Scraping {}".format(browser.title))
    job_title = browser.find_element_by_xpath("//h2[@class='noMargTop margBotXs strong']").text
    company = browser.find_element_by_xpath("//span[@class='strong ib']").text
    city_state = browser.find_element_by_xpath("//span[@class='subtle ib']").text.strip(" â€“ ")
    try:
        browser.find_element_by_xpath("//span[contains(text(),'Location')]").click()
        address = browser.find_element_by_xpath('//*[@id="MapContainer"]/div[1]/p[1]').text
        country_postal = browser.find_element_by_xpath('//*[@id="MapContainer"]/div[1]/p[3]').text
    except NoSuchElementException:
        address = None
        country_postal = None
        pass
    try:
        company_rating = browser.find_element_by_xpath("//span[contains(@class,'compactStars margRtSm')]").text
    except NoSuchElementException:
        company_rating = None
    job_info = {}
    job_info['Title'] = job_title
    job_info['Company'] = company
    job_info['Min Salary'] = None
    job_info['Med Salary'] = None
    job_info['Max Salary'] = None
    job_info['Job ID'] = job_id
    job_info['Adress'] = address
    job_info['Country / Postal'] = country_postal
    job_info['City / State'] = city_state
    job_info['Company Rating'] = company_rating
    return job_info
    
```

```python
error_title = 'Page Not Found | Glassdoor'
page_number = 1
jobs_with_salaries_list = []
all_jobs_list = []
while(True):
    browser.get(url_format.format(page_number))
    if browser.title == error_title:
        break    
    for i in browser.find_elements_by_xpath("//*[contains(@class, 'info infoSalEst infoIcon _ok')]"):
        job_info = {}
        company = i.get_attribute("data-employer-shortname")
        job_title = i.get_attribute("data-jobtitle")
        min_salary = i.get_attribute("data-displayed-min-salary")
        med_salary = i.get_attribute("data-displayed-med-salary") 
        max_salary = i.get_attribute("data-displayed-max-salary")
        job_id = i.get_attribute("data-job-id")
        job_info['Title'] = job_title
        job_info['Company'] = company
        job_info['Min Salary'] = min_salary
        job_info['Med Salary'] = med_salary
        job_info['Max Salary'] = max_salary
        job_info['Job ID'] = job_id
        jobs_with_salaries_list.append(job_info)
        
    all_jobs_urls = {}
    for i in browser.find_elements_by_xpath('//a[contains(@href, "/partner/jobListing")]'):
        url = i.get_attribute("href")
        job_id = url.split("jobListingId=")[1]
        if job_id not in all_jobs_urls:
            all_jobs_urls[job_id]=url    
            
    
    for job_id in all_jobs_urls:
        url = all_jobs_urls[job_id]
        job_info = get_job_info(url, job_id)
        all_jobs_list.append(job_info)
    
                
    page_number = page_number + 1

    
for j1 in all_jobs_list:
        for j2 in jobs_with_salaries_list:
            if j1['Job ID'] == j2['Job ID']:
                j1['Min Salary'] = j2['Min Salary']
                j1['Med Salary'] = j2['Med Salary']
                j1['Max Salary'] = j2['Max Salary']
```

```python
keys = all_jobs_list[0].keys()
output_file_name = input("Please enter the output CSV file name to save: ")
with open(output_file_name, 'w', newline='', encoding='utf-8') as output_file:
    dict_writer = csv.DictWriter(output_file, keys)
    dict_writer.writeheader()
    dict_writer.writerows(all_jobs_list)
```

Displaying jobs locations on the Google Maps

```python
# in order to plot jobs locations on the Google Maps we get the latitudes and longitudes of the addresses
longitude_list = []
latitude_list = []
geolocator = Nominatim(user_agent="sample app")
for address in df['Adress'].values:
    try:
        location = geolocator.geocode(address)
        if location:        
            longitude_list.append(location.longitude)
            latitude_list.append(location.latitude)
        else:
            location = geolocator.geocode(df.loc[df['Adress'] == address, 'City / State'].values[0])
            if location:
                longitude_list.append(location.longitude)
                latitude_list.append(location.latitude)
    except:
        pass

```

```python
# save jobs locations on Google Maps to html file
gmap = gmplot.GoogleMapPlotter(latitude_list[0], longitude_list[0], 4) 
gmap.scatter( latitude_list, longitude_list, 'RED', size = 10000, marker = False ) 
gmap.draw( "jobs_locations.html" )
```
